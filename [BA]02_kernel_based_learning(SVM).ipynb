### 1. Ready to analysis

#### 1) Import packages

```python
import pandas as pd
import numpy as np
import random as rd

import itertools
import os

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore')
```

#### 2) load dataset

```python
# sklearn dataset : breast cancer

breast_cancer = load_breast_cancer()
dir(breast_cancer)
```

```python
# setting X, Y

X = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
Y = pd.DataFrame(breast_cancer.target, columns=["class"])
df = pd.concat([X, Y], axis=1)
print(df.shape)
df
```

#### 3) Exploratory Data Analysis & Pre-process

```python
# Check the y value: to choose a right method for model evaluation
# The imbalance between samples has not been observed, so we can use the method based on confusion matrix

Y.value_counts()
```

```python
# Identifying the properties of features: continuous ot discrete variables, the needs for dummy variables

df.info()
```

```python
for i in df.columns:
    print(df[i].value_counts())
```

```python
# dataset in sklearn package -> saving in json file format
os.chdir('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/data')

df.to_json('breast.json', orient='table')
```

```python
# scaling X set

for i in X.columns:
    scaler.fit(np.array(X[i]).reshape(-1, 1))
    X[i] = scaler.transform(np.array(X[i]).reshape(-1, 1))

X
```

```python
# split train & validate & test set

X_tr, X_tmp, y_tr, y_tmp = train_test_split(X, Y, test_size = 0.4, random_state=1)
X_val, X_te, y_val, y_te = train_test_split(X_tmp, y_tmp, test_size = 0.5, random_state=1)

print(len(X_tr))
print(len(X_val))
print(len(X_te))
print(len(y_tr))
print(len(y_val))
print(len(y_te))
```

### 2. Main Analysis : in whole Variables

#### 1) linear SVM

###### - linear : basic

```python
# fitting

linear = SVC(kernel = "linear", random_state=1)
linear.fit(X_tr, y_tr)
```

```python
# evaluation

linear_pred = linear.predict(X_val)

print("---Confusion Matrix---")
print(confusion_matrix(y_val, linear_pred))
print(" ")
print("---Index for Evaluation---")
print(classification_report(y_val, linear_pred))
```

```python
plot_confusion_matrix(linear,
                      X_val, y_val,                   
                      cmap=plt.cm.Reds, 
                      normalize='true')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/confusion_linear_basic.png', edgecolor='black', format='png', dpi=200)
```

###### - linear : tuning params

```python
# tuning cost : hard & soft margin

# fitting
linear_hard = SVC(kernel = "linear", C=1000, random_state=1)
linear_soft = SVC(kernel = "linear", C=0.01, random_state=1)

linear_hard.fit(X_tr, y_tr)
linear_soft.fit(X_tr, y_tr)
```

```python
# evaluation : compare modified results to original results 

hard_pred = linear_hard.predict(X_val)
print("---Confusion Matrix---")
print(confusion_matrix(y_val, hard_pred))
print(" ")
print("---Index for Evaluation---")
print(classification_report(y_val, hard_pred))

soft_pred = linear_soft.predict(X_val)
print("---Confusion Matrix---")
print(confusion_matrix(y_val, soft_pred))
print(" ")
print("---Index for Evaluation---")
print(classification_report(y_val, soft_pred))

# hard margin : instances in class 1 are classified better than basic but, class 0 are not.
# soft margin : instances in class 0 are classified better than basic but, class 1 are not seriously.
```

```python
plot_confusion_matrix(linear_hard,
                      X_te, y_te, 
                      cmap=plt.cm.Reds,
                      normalize='true')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/confusion_linear_hard.png', edgecolor='black', format='png', dpi=200)

plot_confusion_matrix(linear_soft,
                      X_te, y_te,
                      cmap=plt.cm.Reds,
                      normalize='true')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/confusion_linear_soft.png', edgecolor='black', format='png', dpi=200)
```

```python
# finding optimal hyperparameter and check the result of validation set : C=39~87(find manually) -> setting C=50
linear_opt = SVC(kernel = "linear", C=50, random_state=1)
linear_opt.fit(X_tr, y_tr)

opt_pred = linear_opt.predict(X_val)
print("---Confusion Matrix---")
print(confusion_matrix(y_val, opt_pred))
print(" ")
print("---Index for Evaluation---")
print(classification_report(y_val, opt_pred))
```

```python
# Comparing result of predicting testing set with non-parameter and optimal parameter

plot_confusion_matrix(linear,
                      X_te, y_te,                   
                      cmap=plt.cm.Reds, 
                      normalize='true')

plt.show()

plot_confusion_matrix(linear_opt,
                      X_te, y_te,                   
                      cmap=plt.cm.Reds, 
                      normalize='true')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/confusion_linear_optimal.png', edgecolor='black', format='png', dpi=200)
```

```python
# parameter1 cost : hard & soft -> find appropriate range of cost(manual searching)

gaussian_hard = SVC(kernel = "rbf", C=10, random_state=1)
gaussian_hard.fit(X_tr, y_tr)

gaussian_soft = SVC(kernel = "rbf", C=0.1, random_state=1)
gaussian_soft.fit(X_tr, y_tr)
```

```python
plot_confusion_matrix(gaussian_naro,
                      X_val, y_val, 
                      cmap=plt.cm.Reds,
                      normalize='true')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/confusion_gaussian_naro.png', edgecolor='black', format='png', dpi=200)

plot_confusion_matrix(gaussian_wide,
                      X_val, y_val,
                      cmap=plt.cm.Reds,
                      normalize='true')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/confusion_gaussian_wide.png', edgecolor='black', format='png', dpi=200)
```

### 3. Result Analysis : Visualize dataset with dimentionality reduction

#### 1) Dimentionality reduction : Principal Component Analysis

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
# X_pca & class index matching

index_0 = df[df['class'] == 0].index
index_1 = df[df['class'] == 1].index

X_pca_0 = np.array([X_pca[i] for i in index_0])
X_pca_1 = np.array([X_pca[i] for i in index_1])
```

```python
print(len(index_0))
print(len(X_pca_0))
print(len(index_1))
print(len(X_pca_1))
```

#### 2) Visualization (in 3D space)

```python
X_pca_0 = np.array(X_pca_0).T
X_pca_1 = np.array(X_pca_1).T
print(X_pca_0.shape)
print(X_pca_1.shape)
```

```python
f = plt.figure(figsize=(20,12))
f = plt.axes(projection='3d')
f.scatter(X_pca_0[0], X_pca_0[1], X_pca_0[2], label = 'class_0', c = 'g', s = 50)
f.scatter(X_pca_1[0], X_pca_1[1], X_pca_1[2], label = 'class_1', c = 'r', s = 50)
f.set_xlabel('PCA1')
f.set_ylabel('PCA2')
f.set_zlabel('PCA3')

plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')

plt.savefig('C:/Users/kw764/Desktop/kw/00_lecture/2022-2/BA/튜토리얼이미지/dataset_distribution_1.png', edgecolor='black', format='png', dpi=200)
```
